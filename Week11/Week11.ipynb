```python
# NAME: gradebook.py
# IMPORT STATEMENTS

# FUNCTION DEFINITIONS
# def add_students(grades_dictionary):
def add_students(grades_dictionary):
# create a while loop to input student's name
# when the user type done end the loop
 while True:
  name = input("Enter student's name:")
  if name == 'done':
    break
# Get the grade and convert to an integer
  grade = int(input("Enter grade (0-100): "))
# Set student name as the key and the grade as the value
  grades_dictionary[name] = grade

# def calculate_average(grades_dictionary):
def calculate_average(grades_dictionary):
# Get values from the dictionary
 scores = grades_dictionary.values()
# add all the numbers up
 total_score = sum(scores)
# count how many students there are
 total_students = len(scores)
# total sum divided by total count
# calculate return average
 average = total_score / total_students
 if total_students == 0:
  return 0
 else:
  return average


# PROGRAM LOGIC
# Create an empty dictionary called gradebook
gradebook = {}
# call add_students to put in the dictionary
add_students(gradebook)
#print("Current Gradebook Data:", gradebook)
# call calculate_average and store result
avg = calculate_average(gradebook)
# print final report showing student and grade
print("------Final Report------ ")
for name in gradebook:
 print(name + ": " + str(gradebook[name]))
# print average formatted two decimal places
print("------------")
print("Class Average: " + str(round(avg, 2)))
```
