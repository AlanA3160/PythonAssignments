```python
# NAME: rentalcar.py
# IMPORT STATEMENTS

# FUNCTION DEFINITIONS
#define classification(code)
def classification(code):
#input for classfication code('B', 'S', 'L") return full string ("Budget", "standard", "Luxury")
 if code == 'B':
  return "Budget"
 elif code == 'S':
  return "Standard"
 elif code == 'L':
  return "Luxury"
 else:
   return "Invald"

#define calculate_bill(code, days, miles)
#set daily rate based on class
def calculate_bill(code, days, miles):
  if code == 'B':
    rate = 40
  elif code == 'S':
    rate = 60
  else:
    rate = 80
#Calculate mileage fee(0.35 per miles only for miles over 100)
  chargedmiles = miles - 100
  if chargedmiles < 0:
    chargedmiles = 0
  mileagefee = chargedmiles * 0.35
#Sum the daily cost and mileage cost(return final cost as float)
  total = (rate * days) + mileagefee
  return total
# PROGRAM LOGIC
#1. Input validation
code = ""
while code not in ['B', 'S', 'L']:
  code = input("Enter classification code (B, S, L): ").upper()
  if code not in ['B', 'S', 'L']:
    print("Invalid entry. Use B, S, L ")
#2. User input for odo
startodo = float(input("Starting Odometer: "))
endodo = float(input("Ending Odometer: "))
days = int(input("Days rented: "))
#3. calculations for odo
totalmiles = endodo - startodo
chargedmiles = totalmiles - 100
if chargedmiles < 0:
  chargedmiles = 0
#4. function calls
fullname = classification(code)
totalcost = calculate_bill(code, days, totalmiles)
#5. Final output
print("---RENTAL SUMMARY---")
print("CLASSIFICATION:", fullname)
print("Total miles:", totalmiles)
print("Charged miles:", chargedmiles)
print("Total Cost: $", totalcost)
```
