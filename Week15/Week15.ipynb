```python
import json
import os
from datetime import datetime

class GameToolkit:
    def __init__(self, filename='game_data.json'):
        self.filename = filename
        # Initialize structured data for all three features
        self.data = self.load_all_data()

    def load_all_data(self):
        """Loads all data categories from the JSON file."""
        if os.path.exists(self.filename):
            with open(self.filename, 'r') as file:
                return json.load(file)
        return {
            "backlog": [],
            "inventory": {},  # For Loot & Craft
            "matches": []     # For Meta-Tracker
        }

    def save_all_data(self):
        """Permanent storage for all features."""
        with open(self.filename, 'w') as file:
            json.dump(self.data, file, indent=4)
        print("\n[System] All data synchronized and saved to disk.")

    # --- FEATURE 2: LOOT & CRAFT OPTIMIZER ---
    def manage_crafting(self):
        print("\n--- Loot & Craft Optimizer ---")
        item_name = input("Enter material name (e.g., Iron Ore): ").strip().lower()
        try:
            owned = int(input(f"How many {item_name} do you currently have? "))
            required = int(input(f"How many {item_name} are required for your recipe? "))

            self.data["inventory"][item_name] = {"owned": owned, "required": required}

            diff = required - owned
            if diff > 0:
                print(f">> You still need to farm {diff} more {item_name}.")
            else:
                print(f">> You have enough! Surplus: {abs(diff)}")

            self.save_all_data()
        except ValueError:
            print("Error: Please enter whole numbers for item counts.")

    # --- FEATURE 3: META-TRACKER (STAT ANALYZER) ---
    def log_match(self):
        print("\n--- Meta-Tracker Match Log ---")
        character = input("Character Played: ")
        result = input("Result (Win/Loss): ").strip().capitalize()

        if result not in ["Win", "Loss"]:
            print("Invalid result. Please enter 'Win' or 'Loss'.")
            return

        match_entry = {
            "timestamp": datetime.now().strftime("%Y-%m-%d %H:%M"),
            "character": character,
            "result": result
        }

        self.data["matches"].append(match_entry)
        self.calculate_win_rates()
        self.save_all_data()

    def calculate_win_rates(self):
        """Logic: Calculates performance percentages per character."""
        if not self.data["matches"]:
            print("No match history found.")
            return

        stats = {}
        for m in self.data["matches"]:
            char = m["character"]
            if char not in stats:
                stats[char] = {"wins": 0, "total": 0}
            stats[char]["total"] += 1
            if m["result"] == "Win":
                stats[char]["wins"] += 1

        print("\n--- Current Win Rates ---")
        for char, data in stats.items():
            win_rate = (data["wins"] / data["total"]) * 100
            print(f"{char}: {win_rate:.1f}% win rate ({data['wins']}/{data['total']})")

    # --- RE-USING BACKLOG FEATURE ---
    def add_backlog_game(self):
        print("\n--- Add Game to Backlog ---")
        title = input("Game Title: ")
        try:
            hours = float(input("Hours to beat: "))
            self.data["backlog"].append({"title": title, "hours": hours})
            self.save_all_data()
        except ValueError:
            print("Error: Please enter a valid number for hours.")

def main():
    toolkit = GameToolkit()

    while True:
        print("\n=== ULTIMATE GAMING TOOLKIT ===")
        print("1. Backlog: Add Game")
        print("2. Crafting: Track Materials")
        print("3. Meta-Tracker: Log Match")
        print("4. Exit")

        choice = input("\nSelect an option: ")

        if choice == '1':
            toolkit.add_backlog_game()
        elif choice == '2':
            toolkit.manage_crafting()
        elif choice == '3':
            toolkit.log_match()
        elif choice == '4':
            print("Closing program. Data is safe!")
            break

if __name__ == "__main__":
    main()
```
