```python
# NAME: journal.py
# IMPORT STATEMENTS

# FUNCTION DEFINITIONS
# Function: write_entry(filename, entry)
def write_entry(filename, entry):
# Open file 'fout' in append mode ('a')
 fout = open(filename, 'a')
# Write to the file and add a new line
 fout.write(entry + '\n')
# Close the file
 fout.close()
# Write a success message
print("Entry Saved")

#Function: read_journal(filename)
def read_journal(filename):
# use try/except for FileNotFoundError
 try:
# open file 'fin' in read mode ('r')
  fin = open(filename, 'r')
# Print journal content and close
  content = fin.read()
  print("\n --- Your Journal --- ")
  print(content)
  fin.close()
 except FileNotFoundError:
  print("\nError: no journal found!")

# PROGRAM LOGIC
# Intialize the journal filename variable
journal_filename = 'journal.txt'
# Create a while true loop asking user if they want to (W)rite or (R)ead, or (Q)uit Format case
while True:
 choice = input("\n(W)rite, (R)ead or (Q)uit").upper()
# If choice is W, ask user for their thoughts and call write entry
 if choice == 'W':
  entry = input("Enter your thoughts:")
  write_entry(journal_filename, entry)
# If choice is R, call read_journal
 elif choice == 'R':
   read_journal(journal_filename)
# If choice is Quit Print Goodbye
 elif choice == 'Q':
  print("Closing..GOODBYE")
  break
```
