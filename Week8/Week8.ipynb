```python
# IMPORT STATEMENTS

# FUNCTION DEFINITIONS
# Define a function named is_palindrome that accepts one parameter
def is_palindrome(word):
# convert word to lower case
  first_word = word.lower()
# create reversed version of the word
  reversed_word = first_word[::-1]
# compare lower case word to reversed word
  return first_word == reversed_word
# return true if they match, else return false


# PROGRAM LOGIC
# 1. ask user to type in a word and save it to a variable
user_input = input("Enter a word to test: ")
# 2. call the is_palindrome function and pass as argument
result = is_palindrome(user_input)
# 3. if the function returns true print, else print
if result == True:
  print(f'{user_input}' " is a palindrome")
else:
  print(f'{user_input}' " is not a palindrome")
```
