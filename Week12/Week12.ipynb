```python
# NAME: rectangle.py
# IMPORT STATEMENTS


# FUNCTION DEFINITIONS
 # Define a function that takes length * width that returns both area and perimeter
 # def calculate_rectangle length * width: return area, perimeter
def calculate_rectangle(length, width):
 # calculate area
  area = length * width
 # calculate perimeter
  perimeter = 2 * (length + width)
 # return both values
  return area, perimeter


# PROGRAM LOGIC
# 1. Ask the user for length and width
# 2. Unpack the returned tuple into separate variables
# 3. Print the results
print(calculate_rectangle(5, 10))
```
