```python
# NAME: roster.py
# IMPORT STATEMENTS

# FUNCTION DEFINITIONS
# Define add_players(player_list):
def add_players(players_list):
#Start a while loop untill user types quit
  while True:
    name = input("Enter a player name (or type quit to stop): ")
#If input is not quit, append name to player_list
    if name == "quit":
      break
    players_list.append(name)
  return players_list


# Define is_on_roster(player_list, search_name)
def is_on_roster(player_list, search_name):
# Check if search_name is in player_list
  if search_name in player_list:
    return True
  else:
    return False
# Return true or false

# PROGRAM LOGIC
# 1. create empty roster list
roster = []
# 2. call add_players
add_players(roster)
# 3. Get search name from user
search_name = input("Enter player name to search ")
# 4. Call is_on_roster and print result
#print("Current roster:", roster)
if is_on_roster(roster, search_name):
  print("Player found!")
else:
  print("player not found.")
```
